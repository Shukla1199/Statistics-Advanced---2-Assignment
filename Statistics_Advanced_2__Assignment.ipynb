Q6-Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.
(Include your Python code and output in the code box below.)

To generate a binomial distribution and plot its histogram in Python, you can use the numpy library to generate the random numbers and the matplotlib library for plotting. The numpy.random.binomial function generates samples based on the number of trials (n) and the probability of success (p)

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Generate the binomial distribution data ---
n = 10  # Number of trials
p = 0.5 # Probability of success
size = 1000 # Number of random variates to generate (sample size for visualization)

# Generate random numbers from a binomial distribution
# Each number represents the count of successes across 'n' trials
# A larger 'size' gives a better representation of the distribution shape
data_binomial = np.random.binomial(n, p, size)

# --- 2. Plot the histogram ---

# Use the 'hist' function from matplotlib.pyplot to plot the data
plt.hist(data_binomial, bins=range(n + 2), density=True, edgecolor='black', rwidth=0.8)

# Add labels and a title for clarity
plt.title('Binomial Distribution Histogram (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.grid(axis='y', alpha=0.75) # Optional: Add gridlines for better readability

# Display the plot
plt.show()
```

Q7-Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,

This analysis performs a one-sample Z-test (
) to determine if the sample mean significantly differs from a assumed population mean of
 at a
 significance level (
). The Z-statistic is calculated as
. Assuming a population standard deviation
, we fail to reject the null hypothesis, indicating no significant difference

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

# Parameters
null_mean = 50.0
# Assuming population standard deviation (sigma) is 0.5
pop_std = 0.5
alpha = 0.05

# 1. Calculate Sample Statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)
std_error = pop_std / np.sqrt(n)

# 2. Calculate Z-statistic
z_stat = (sample_mean - null_mean) / std_error

# 3. Calculate P-value (Two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# 4. Interpretation
if p_value < alpha:
    print("Reject the null hypothesis (Significant difference)")
else:
    print("Fail to reject the null hypothesis (No significant difference)")
```

Q8-Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


The following Python script uses numpy to simulate data from a normal distribution, scipy.stats to calculate the 95% confidence interval for the mean, and matplotlib to plot the data.

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# 1. Simulate data from a normal distribution
np.random.seed(42) # for reproducibility
mu, sigma, sample_size = 50, 5, 1000 # Mean, standard deviation, and sample size
simulated_data = np.random.normal(mu, sigma, sample_size)

# 2. Calculate the 95% confidence interval for the mean
sample_mean = np.mean(simulated_data)
# Calculate standard error of the mean (SEM)
# When the sample size is large (n > 30), the t-distribution approximates the normal distribution.
# We use st.t.interval, which is suitable for both small and large samples with unknown population std dev.
confidence_interval = st.t.interval(
    confidence=0.95,
    df=len(simulated_data)-1,
    loc=sample_mean,
    scale=st.sem(simulated_data)
)

# Print the results
print(f"Simulated Data Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
print(f"True Population Mean (for reference): {mu}")

# 3. Plot the data using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(simulated_data, bins=30, density=True, alpha=0.7, color='skyblue', label='Simulated Data Histogram')
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Sample Mean: {sample_mean:.2f}')

# Plot the confidence interval lines
plt.axvline(confidence_interval[0], color='green', linestyle='dotted', linewidth=2, label=f'95% CI Lower Bound: {confidence_interval[0]:.2f}')
plt.axvline(confidence_interval[1], color='green', linestyle='dotted', linewidth=2, label=f'95% CI Upper Bound: {confidence_interval[1]:.2f}')

# Add a line for the true population mean
plt.axvline(mu, color='orange', linestyle='solid', linewidth=2, label=f'True Mean: {mu}')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution Simulation with Mean and 95% Confidence Interval')
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.show()
```

The following Python function uses the NumPy and Matplotlib libraries to calculate Z-scores and visualize the standardized data with a histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def analyze_standardized_data(data):
    """
    Calculates Z-scores for a dataset and visualizes the standardized data
    using a histogram.

    Args:
        data (array-like): The input dataset (e.g., a list or NumPy array).
    """

    # Calculate Z-scores using scipy.stats.zscore for efficiency
    # The 'ddof=0' parameter specifies the population standard deviation
    # calculation, consistent with the standard z-score formula (X - mu) / sigma
    z_scores = zscore(data, ddof=0)

    # Optional: Manual calculation (alternative to scipy.stats.zscore)
    # mean = np.mean(data)
    # std_dev = np.std(data)
    # z_scores = (data - mean) / std_dev

    print("Original Data:", data)
    print("Z-Scores:", z_scores)
    print(f"Mean of Z-scores: {np.mean(z_scores):.2f}")
    print(f"Standard Deviation of Z-scores: {np.std(z_scores):.2f}")

    # Visualize the standardized data using a histogram
    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=20, color='skyblue', edgecolor='black', density=True)
    plt.title("Histogram of Standardized Data (Z-Scores)")
    plt.xlabel("Z-Score (Standard Deviations from Mean)")
    plt.ylabel("Density")
    plt.grid(True)
    plt.show()

# Example usage with sample data:
sample_data = np.array([6, 7, 7, 12, 13, 13, 15, 16, 19, 22])
analyze_standardized_data(sample_data)
```
